# pie chart

```python
counts = df['room_type'].value_counts()
labels = counts.index.tolist()  # Get the labels from the index of value_counts()

colors = ['lavender', 'white','pink','orange']

counts = counts.sort_index(ascending=True)
wedgeprops = {'linewidth': 10, 'edgecolor': 'black'}

# Adjust the figure size here
plt.figure(figsize=(10, 10))  

plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Set the title on the graph
plt.title("room_type", color='grey', loc='left', fontweight='1', fontsize=8)

plt.legend(labels, loc="upper right")

plt.show()
```
